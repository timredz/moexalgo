```python
from datetime import datetime, timedelta
from moexalgo import shares

moex = shares.get('MOEX')
moex_candles_df = moex.candles(datetime.now()-timedelta(hours=3), interval=1)
moex_candles_df
```

```python
import mplfinance as mpf
mpf.plot(moex_candles_df.set_index('begin'), type='candle', style='yahoo', volume=True)
```
